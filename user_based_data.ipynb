# 先計算 user 瀏覽紀錄的產業觀看次數

```python
before_index = -1
step = 0
user_industry_prefer_count_list = []
############################ 按照 歷史瀏覽資料的次數找出最後一筆的 index  
for counts in id_count_final.counts:
    index = counts + before_index
############################ 透過將該 index 逐步減去 range(歷史瀏覽資料的次數)  
    user_industry_prefer_count = 0
    for t in range(counts):
        tem = final_final.iloc[index - t ,3:] # tem 為該 user 的產業 list
        
        user_industry_prefer_count = user_industry_prefer_count + tem
        
        
    tem_li = user_industry_prefer_count.tolist()
    user_industry_prefer_count_list.append(tem_li)  
    
    
    before_index = index
```

```python
user_industry_prefer_count = pd.DataFrame(user_industry_prefer_count_list)
```

```python
user_industry_prefer_count.head()
```

```python
user_industry_prefer_count.columns = final_final.columns[3:]
```

```python
user_industry_prefer_count.head(3)
```

```python
user_industry_prefer_data = pd.concat([id_count_final.user_id, user_industry_prefer_count], axis=1)
```

```python
user_industry_prefer_data.head(3)
```

```python
user_industry_prefer_data.to_csv('user_industry_prefer_data.csv',index=0)
```

```python
user_industry_prefer_data=pd.read_csv('user_industry_prefer_data.csv')
user_industry_prefer_data.head(3)
```

```python
n = 5
a = 0.3
```

```python
batch_size=10000
li = []
while(len(user_industry_prefer_data)>0):
    
    print('剩',len(user_industry_prefer_data),'還沒處理')
    temp = user_industry_prefer_data.iloc[:,1:].head(batch_size)
                
               
    for index, row in tqdm_notebook(temp.iterrows()):
        prefer = []
        for i in range(895):
            if row[i] >= n:
                prefer.append(1)
            elif ( row[i]/id_count_final.counts[index] ) >= a:
                prefer.append(1)
            else:
                prefer.append(0)

        li.append(prefer)           

        
    if len(user_industry_prefer_data) > batch_size:
        #把已經處理過的資料drop掉
        user_industry_prefer_data = user_industry_prefer_data.drop(np.arange(batch_size))
        id_count_final = id_count_final.drop(np.arange(batch_size))
        #重設index，drop是by index
        user_industry_prefer_data.reset_index(drop=True,inplace=True)
        id_count_final.reset_index(drop=True,inplace=True)
    else:
        break    
```

```python
# 因為前面處理資料用批次處理的方式，所以這裡重新 load data 回來
user_industry_prefer_data = pd.read_csv('user_industry_prefer_data.csv')
```

```python
final_industry_prefer_data = pd.DataFrame(li)
final_industry_prefer_data.columns = user_industry_prefer_data.columns[1:]
```

```python
final_industry_prefer_data.head(3)
```

```python
uesr_based_data = pd.concat([user_industry_prefer_data.iloc[:,0], final_industry_prefer_data], axis=1)
```

```python
uesr_based_data.head(3)
```

```python
uesr_based_data.to_csv('user_based_data.csv',index=0)
```
